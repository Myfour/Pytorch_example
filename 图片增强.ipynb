```python
'''
os.walk的结果为一个3个值的元组（dirpath,dirnames,filenames）
dirpath读取到的每一个跟目录
dirnames每个根目录下的文件夹有哪些
fillenames该根目录下的所有文件名称
'''
for root,dirs,files in os.walk(r'/home/nifer/Desktop/图片集'):                                                 
    print(root)
    print('-----------------------------------------------')
    print(dirs)
    print('-----------------------------------------------')
    print(files)
    print('-----------------------------------------------')
```

```python
'''
得到完整文件路径名
'''
filepaths=[]
for root,dirs,files in os.walk(r'/home/nifer/Desktop/图片集'):
    for filename in files:
        print(os.path.join(root,filename))  # os.path.join()用来拼接一个文件路径
        filepaths.append(os.path.join(root,filename))
```

```python
'''
对图像进行4个90度的旋转处理
'''
for file in filepaths:
    im=Image.open(r''+file)
    dirName,fileFullName=os.path.split(file)  # 提取文件夹和文件全名
    fileName,extName=os.path.splitext(fileFullName)  # 提出文件扩展名
    
    rotate90=im.rotate(90)
#     print(os.path.join(dirName,fileName+'_rotate90'+extName))
    rotate90.save(r''+os.path.join(dirName,fileName+'_rotate90'+extName))
    rotate180=im.rotate(180)
    rotate180.save(r''+os.path.join(dirName,fileName+'_rotate180'+extName))
    rotate270=im.rotate(270)
    rotate270.save(r''+os.path.join(dirName,fileName+'_rotate270'+extName))
    
```

```python
'''
对图像进行水平翻转处理
'''
for file in filepaths:
    im=Image.open(r''+file)
    dirName,fileFullName=os.path.split(file)  # 提取文件夹和文件全名
    fileName,extName=os.path.splitext(fileFullName)  # 提出文件扩展名
    im.transpose(Image.FLIP_LEFT_RIGHT).save(r''+os.path.join(dirName,fileName+'_LR'+extName))
```

```python
'''
对图像进行锐化、细节提取、轮廓提取、深度平滑处理
'''
for file in filepaths:
    im=Image.open(r''+file)
    dirName,fileFullName=os.path.split(file)  # 提取文件夹和文件全名
    fileName,extName=os.path.splitext(fileFullName)  # 提出文件扩展名
    if im.mode=='RGB':
        im.filter(ImageFilter.CONTOUR).save(r''+os.path.join(dirName,fileName+'_CONTOUR'+extName))  # 轮廓边缘提取过滤器
        im.filter(ImageFilter.DETAIL).save(r''+os.path.join(dirName,fileName+'_DETAIL'+extName))  # 细节过滤器
        im.filter(ImageFilter.SHARPEN).save(r''+os.path.join(dirName,fileName+'_SHARPEN'+extName))  # 锐化过滤器
        im.filter(ImageFilter.SMOOTH_MORE).save(r''+os.path.join(dirName,fileName+'_SMOOTH_MORE'+extName))  # 深度平滑过滤器
```
